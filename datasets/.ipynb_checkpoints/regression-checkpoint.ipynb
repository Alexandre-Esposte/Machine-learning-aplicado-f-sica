# Model Selection -  Cross Validation


```python
# importar pacotes
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
from sklearn import linear_model


# Configura o número de linhas e colunas que serão mostrados, e o comprimento das colunas nos dataframes
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('max_colwidth', None)

# para conseguir repetir o experimento 
seed = 50
np.random.seed(seed)
```

```python
df = pd.read_csv('AB2_ehull_materials_spacegroup2.csv')
df.drop('Unnamed: 0', axis='columns', inplace=True)
df.head()
```

# Checking ehull distribution


```python
hform = df['ehull'].copy()
plt.hist(hform, bins=8, density=False, label='Data', rwidth=0.9)
```

# Check for missing entries

# Train / Test dataset spliting

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.1, shuffle=True, random_state=seed)
```

```python
train_set.head()
```

```python
test_set.head()
```

# Training set

```python
X = train_set.drop(['ehull', 'Material'], axis='columns') #(Features)
y = train_set['ehull'] #(Target)
print(X.shape,y.shape)
X.head()
```

```python
X_test = test_set.drop(['ehull', 'Material'], axis='columns') #(Features)
y_test = test_set['ehull'] #(Target)
print(X_test.shape,y_test.shape)
```

# Least Square Methdod

```python
from sklearn import metrics
from sklearn.models

model_lsm = LinearRegression() # least square minimum
model_lsm.fit(X,y)

y_pred_train = model_lsm.predict(X)
train_error  = metrics.mean_squared_error(y, y_pred_train, squared=True)


y_pred_test=model_lsm.predict(X_test)
test_error=metrics.mean_squared_error(y_test, y_pred_test, squared=True)

print(train_error,test_error)
```

```python
print(model_lsm.coef_)
print(np.max(np.abs(model_lsm.coef_)),np.min(np.abs(model_lsm.coef_)))
print(np.argmax(model_lsm.coef_),np.argmin(np.abs(model_lsm.coef_)))

```

# Ridge Regression

```python
from sklearn import metrics

model_rdg = linear_model.Ridge(alpha=15) # least square minimum
model_rdg.fit(X,y)

y_pred_train = model_rdg.predict(X)
train_error  = metrics.mean_squared_error(y, y_pred_train, squared=True)


y_pred_test=model_rdg.predict(X_test)
test_error=metrics.mean_squared_error(y_test, y_pred_test, squared=True)

print(train_error,test_error)
```

```python
print(model_rdg.coef_)
print(np.max(np.abs(model_rdg.coef_)),np.min(np.abs(model_rdg.coef_)))
print(np.argmax(model_rdg.coef_),np.argmin(np.abs(model_rdg.coef_)))
```

# Lasso

```python
from sklearn import metrics

model_lasso = linear_model.Lasso(alpha=1,max_iter=100000) # least square minimum
model_lasso.fit(X,y)

y_pred_train = model_lasso.predict(X)
train_error  = metrics.mean_squared_error(y, y_pred_train, squared=True)


y_pred_test=model_lasso.predict(X_test)
test_error=metrics.mean_squared_error(y_test, y_pred_test, squared=True)

print(train_error,test_error)
```

```python
print(model_lasso.coef_)
print(np.max(np.abs(model_lasso.coef_)),np.min(np.abs(model_lasso.coef_)))
print(np.argmax(model_lasso.coef_),np.argmin(np.abs(model_lasso.coef_)))
```

# Tradition Cross Validation

```python
from IPython.display import Image
Image(filename='validacao_cruzada_trad.png', width=500, height=500)
```

```python
from sklearn.model_selection import cross_val_score

train_set, test_set = train_test_split(df, test_size=0.2, shuffle=True, random_state=seed) 

X_train = train_set.drop(['ehull', 'Material'], axis='columns')
y_train = train_set['ehull'] 

X_test = test.drop(['ehull', 'Material'], axis='columns')
y_test = test_set['ehull'] 


model_lsm = LinearRegression() # least square minimum

model_rdg = linear_model.Ridge(alpha=1) # least square minimum

model_lasso = linear_model.Lasso(alpha=1,max_iter=100000) # least square minimum

scores_lsm = cross_val_score(model_lsm, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
scores_rdg = cross_val_score(model_rdg, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
scores_lasso = cross_val_score(model_lasso, X_train, y_train, cv=5, scoring='neg_mean_squared_error')

print("------------MSE each fold----------------\n")

print("LSM       = ",scores_lsm*-1)
print("Ridge Reg = ",scores_rdg*-1)
print("Lasso Reg = ",scores_lasso*-1)

```

```python
print("------------Averaged MSE----------------\n")
print("LSM       = ",scores_lsm.mean()*-1)
print("Ridge Reg = ",scores_rdg.mean()*-1)
print("Lasso Reg = ",scores_lasso.mean()*-1)
```

```python
from sklearn.preprocessing import PolynomialFeatures


train_set, test_set = train_test_split(df, test_size=0.2, shuffle=True, random_state=seed) 

X_train = train_set.drop(['ehull', 'Material'], axis='columns')
y_train = train_set['ehull'] 

X_test = test.drop(['ehull', 'Material'], axis='columns')
y_test = test_set['ehull'] 


# ------------------- Training ---------------------
poly = PolynomialFeatures(degree=2,include_bias=False)
X_poly=poly.fit_transform(X_train)

# LSM
model_lsm.fit(X_poly, y_train)
y_train_pred_lsm = model_lsm.predict(X_poly)

# Ridge
model_rdg.fit(X_poly, y_train)
y_train_pred_rdg = model_rdg.predict(X_poly)

# Lasso
model_lasso.fit(X_poly, y_train)
y_train_pred_lasso = model_lasso.predict(X_poly)

lsm_train=metrics.mean_squared_error(y_train, y_train_pred_lsm) 
rdg_train=metrics.mean_squared_error(y_train, y_train_pred_rdg) 
lasso_train=metrics.mean_squared_error(y_train, y_train_pred_lasso) 



print("MSE--- LSM=",lsm_train,"--- Ridge Reg=",rdg_train,"--- Lasso Reg=",lasso_train)
```

```python
# ------------------- Test ---------------------
poly = PolynomialFeatures(degree=2,include_bias=False)
X_poly=poly.fit_transform(X_test)

# LSM
model_lsm.fit(X_poly, y_test)
y_test_pred_lsm = model_lsm.predict(X_poly)

# Ridge
model_rdg.fit(X_poly, y_test)
y_test_pred_rdg = model_rdg.predict(X_poly)

# Lasso
model_lasso.fit(X_poly, y_test)
y_test_pred_lasso = model_lasso.predict(X_poly)

lsm_test=metrics.mean_squared_error(y_test, y_test_pred_lsm) 
rdg_test=metrics.mean_squared_error(y_test, y_test_pred_rdg) 
lasso_test=metrics.mean_squared_error(y_test, y_test_pred_lasso) 

print("MSE--- LSM=",lsm_test,"--- Ridge Reg=",rdg_test,"--- Lasso Reg=",lasso_test)
```
