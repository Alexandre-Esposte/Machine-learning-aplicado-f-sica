```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
paths = pd.read_csv('../datasets/projeto_final/paths.csv')
```

```python
paths
```

```python
spec = pd.read_csv(paths.iloc[12,0])
```

```python
spec.columns= ['wavenumber','intensity']
```

```python
fig, ax = plt.subplots(figsize=(16,8))


ax.plot(spec['wavenumber'],spec['intensity'],linewidth=1.5,c='blue',label='Função')

ax.set_xlabel('Número de onda (cm-1)',size=13)
ax.set_ylabel('Intensidade (Unidade arbitrária)',size=13)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

spec_aux = spec.query('wavenumber < 5690 and wavenumber > 5682')


axin = ax.inset_axes([0.1, 0.52, 0.45, 0.47])
axin.plot(spec_aux['wavenumber'],spec_aux['intensity'],linewidth=1.5,c='blue')
axin.set_xticks([])
axin.set_yticks([])
ax.indicate_inset_zoom(axin)

fig.savefig('imgs/Espectro.png',dpi=500)
```
